##Operaciones con conjuntos

union(set): Realiza la operacion “union” entre dos conjuntos. La unión entre

1.   union(set): Realiza la operacion “union” entre dos conjuntos. La unión entre dos conjuntos es sumar los elementos de estos sin repetir elementos. Esta operación tambien se puede realizar con el signo “|”: set_a | set_b.
2.   intersection(set): Realiza la operacion “intersection” entre dos conjuntos. La intersección entre dos conjuntos es tomar unicamente los elementos en común de los conjutnos. Esta operación tambien se puede realizar con el signo “&”: set_a & set_b.
3.  difference(set): Realiza la operacion “difference” entre dos conjuntos. La diferencia entre dos conjuntos es restar los elementos del segundo conjunto al primero. Esta operación tambien se puede realizar con el signo “-”: set_a - set_b.
4.  symmetric_difference(set): Realiza la operacion “symmetric_difference” entre dos conjuntos. La diferencia simetrica entre dos conjutnos consta de restar todos los elementos de ambos exceptuando el elemento en común. Esta operación tambien se puede realizar con el signo “^”: set_a ^ set_b.


```python
set_a = {'col', 'mex', 'bol'}
set_b = {'pe', 'bol'}

set_c = set_a.union(set_b)
print(set_c)
print(set_a | set_b)

set_c = set_a.intersection(set_b)
print(set_c)
print(set_a & set_b)

set_c = set_a.difference(set_b)
print(set_c)
print(set_a - set_b)

set_c = set_a.symmetric_difference(set_b)
print(set_c)
print(set_a ^ set_b)
```

##Refactor Game: Piedra, Papel o Tijera

```python
import random

print("[ 🤖 Bienvenido al juego Piedra, Papel o tijera 🙋] >>> Ingresa una opción <<<\n")
result = {
          'piedra' : { 'tijera' : 1, 'papel' : 0 },
          'papel' : { 'tijera' : 0, 'piedra' : 1 },
          'tijera' : { 'papel' : 1, 'piedra' : 0 }
}
def choose_options():
    options = ('piedra', 'papel', 'tijera')
    user_option = input('>>> Piedra, papel o tijera => ').lower()

    if not user_option in options:
        print('Esa opción no es valida')
        #continue
        return None, None

    computer_option = random.choice(options)

    print('User option => ', user_option)
    print('Computer option => ', computer_option)
    return user_option, computer_option


def check_rules(user_option, computer_option, user_wins, computer_wins):
    if user_option == computer_option:
        print('Empate!\n')

    else:
      ganador = result[user_option][computer_option]
      if ganador:
          print("{} gana a {}".format(user_option, computer_option))
          print('¡User gana!\n')
          user_wins += 1
      else:
          print("{} gana a {}".format(computer_option,user_option))
          print('¡Computer gana!\n')
          computer_wins += 1

    return user_wins, computer_wins


def check_winner(user_wins, computer_wins):

    print(f'''
    🤖 Computer wins: {computer_wins}
    🙋 User wins: {user_wins}
            ''')

    if user_wins == 3:
        print('🎖️ El ganador es User 🎖️')
        return False

    if computer_wins == 3:
        print('🎖️ El ganador es Computer 🎖️')
        return False

    return True


def run_game():
    computer_wins = 0
    user_wins = 0
    rounds = 1

    while True:
        print('***' * 10)
        print('Round ', rounds)
        print('***' * 10)

        rounds += 1
        user_option, computer_option = choose_options()
        user_wins, computer_wins = check_rules(user_option, computer_option,user_wins, computer_wins)
        if not check_winner(user_wins, computer_wins):
            break
run_game()
```

##Funciones normales vs Funciones lambda

####Funciones normales (definidas por el usuario):

1.  Uso general: Las funciones normales son adecuadas para tareas más grandes y complejas que requieren varias líneas de código y pueden tener un nombre descriptivo.
2.  Reusabilidad: Si necesitas utilizar la misma lógica en múltiples partes de tu código, es mejor definir una función normal que puedas llamar en varios lugares.
3.  Legibilidad: Las funciones normales suelen ser más legibles y claras, ya que puedes agregar comentarios y dividir el código en bloques más pequeños.
4.  Funcionalidades más avanzadas: Las funciones normales pueden contener múltiples expresiones, sentencias de control (if, while, etc.), bucles y cualquier otro tipo de operación que necesites para realizar una tarea compleja.

####Funciones lambda (funciones anónimas):
1.  Funciones simples y cortas: Las funciones lambda son útiles cuando necesitas una función pequeña y simple que realice una operación rápida y sin la necesidad de darle un nombre.
2.  Funciones de una sola expresión: Las funciones lambda están limitadas a una sola expresión y no pueden contener múltiples instrucciones.
3.  Funciones como argumentos: Las funciones lambda son útiles cuando necesitas pasar una función como argumento a otra función, como en el caso de map(), filter(), sorted(), etc.

```python
#function normal
def increment(x):
  return x + 1

#lambda function
increment_v2 = lambda x: x + 1

result = increment(10)
print(result)

result = increment_v2(20)
print(result)

full_name = lambda name, last_name: f'Full name is {name.title()} {last_name.title()}'

text = full_name('Sebastian', 'Castiblanco Cuvides')
print(text)
```

##Higher order function: una función dentro de otra función (HOF)

```python
def increment(x):
  return x + 1

increment_v2 = lambda x: x +1

#Resive dos parametros un argumento y una función
def high_ord_func(x, func):
  return x + func(x)

high_ord_func_v2 = lambda x, func: x + func(x)

result = high_ord_func(2, increment)
# 2 + (2 + 1)
print(result)

result = high_ord_func_v2(2, increment_v2)
print(result)
result = high_ord_func_v2(2, lambda x: x + 2)
print(result)
## change
result = high_ord_func_v2(2, lambda x: x * 5)
print(result)
```

##Map()

Cuando deseas aplicar una función a cada elemento de una lista y obtener una nueva lista con los resultados. Por ejemplo, para transformar una lista de números en una lista de sus cuadrados.

```python
numbers = [1, 2, 3, 4]
numbers_v2 = []
for i in numbers:
  numbers_v2.append(i * 2)

#map(funcion,lista)
numbers_v3 = list(map(lambda i: i * 2, numbers))

print(numbers)
print(numbers_v2)
print(numbers_v3)

numbers_1 = [1, 2, 3, 4]
numbers_2 = [5, 6, 7]

print(numbers_1)
print(numbers_2)
result = list(map(lambda x, y: x + y, numbers_1, numbers_2))
print(result)
```

##Map con diccionarios

```python
items = [
  {
    'product': 'camisa',
    'price': 100,
  },
  {
    'product': 'pantalones',
    'price': 300
  },
  {
    'product': 'pantalones 2',
    'price': 200
  }
]

prices = list(map(lambda item: item['price'], items))
print(items)
print(prices)

def add_taxes(item):
  item['taxes'] = item['price'] * .19
  return item

#Modifica la lista original
new_items = list(map(add_taxes, items))
print(new_items)
print(items)
```

##Reto: map con inmutabilidad

```python
items = [
  {
    'product': 'camisa',
    'price': 100,
  },
  {
    'product': 'pantalones',
    'price': 300
  },
  {
    'product': 'pantalones 2',
    'price': 200
  }
]

def add_taxes(item):
  #Saca la referencia de la memoria para que modifique el diccionario original
  new_item = item.copy()
  new_item['taxes'] = new_item['price'] * .19
  return new_item

new_items = list(map(add_taxes, items))
print('New list')
print(new_items)
print('Old list')
print(items)
```

##Reduce()

Acumula los elementos de una secuencia aplicando una función de dos argumentos acumulativa. Toma como argumento un conjunto de valores (una lista, una tupla, o cualquier objeto iterable) y lo “reduce” a un único valor. ¿Cómo se obtiene ese único valor?, Se obtiene a partir de la colección pasada como argumento dependerá de la función aplicada.

```python
import functools

numbers = [1, 2, 3, 4]

def accum(counter, item):
  print('counter => ',counter)
  print('item => ',item)
  return counter + item
result = functools.reduce(accum, numbers)
print(result)

result = functools.reduce(lambda counter, item: counter + item, numbers)
print(result)
```

##Manejo de excepciones

Las excepciones son errores que se disparan durante la ejecución de un programa. En Python podemos dejar que dichas excepciones detengan el programa o en caso contrario escribir un algoritmo para reaccionar a dicha situación.

Los ejemplos más comunes que podemos nombrar de excepciones:

*   Tratar de convertir a entero un string que no contiene valores numéricos.
*   Tratar de dividir por cero.
*   Abrir un archivo de texto inexistente o que se encuentra bloqueado por otra aplicación.
*   Conectar con un servidor de bases de datos que no se encuentra activo.
*   Acceder a subíndices de listas o tuplas inexistentes.
*   Llamada a un método o función inexistente.
*   Importar módulos que no existen.

La captura de excepciones nos permite crear programas mucho más robustos y tolerante a fallas que ocurren en escasas situaciones, pero en caso que se presenten disponemos de un algoritmo alternativo para reaccionar a dicha situación evitando que el programa finalice su ejecución.

```python
try:
  print(0 / 0)
  assert 1 != 1, 'Uno no es igual que uno'
  age = 10
  if age < 18:
    raise Exception('No se permiten menores de edad')
except ZeroDivisionError as error:
  print(error)
except AssertionError as error:
  print(error)
except Exception as error:
  print(error)

print('Hola')
print('Hola 2')
```

```python
try:
    # Código que podría generar una excepción
    resultado = 10 / 0  # Genera una división por cero (ZeroDivisionError)
except Exception as e:
    # Manejo de la excepción
    print("Ocurrió una excepción:", e)
    #raise  # Vuelve a lanzar la excepción (opcional)
else:
    # Código que se ejecuta si no se generó ninguna excepción
    print("El resultado es:", resultado)
finally:
    # Código que se ejecuta siempre, ya sea con o sin excepción
    print("Este bloque siempre se ejecuta")

# Salida:
# Ocurrió una excepción: division by zero
# Este bloque siempre se ejecuta
```

```python
#main
#%%writefile main.py
import utils
import read_csv
import charts

def run():
  data = read_csv.read_csv('/content/data.csv')
  country = input('Type Country => \n')

  result = utils.population_by_country(data, country)

  if len(result) > 0:
    country = result[0]
    labels, values = utils.get_population(country)
    charts.generate_bar_chart(labels, values)

#Entry point y que tenga dualidad, con esto se ejecuta desde example y desde main con la terminal
if __name__ == '__main__':
  run()
```

#Graficando el porcentaje de la población de un continente


```python
#main
#%%writefile main.py
import utils
import read_csv
import charts

def run():
  data = read_csv.read_csv('/content/data.csv')
  #Filtra por continente
  data = list(filter(lambda item : item['Continent'] == 'South America',data))
  #Obtiene la lista de paises por el continente
  countries = list(map(lambda x: x['Country'], data))
  #Obtiene la lista de porcentajes
  percentages = list(map(lambda x: x['World Population Percentage'], data))
  charts.generate_pie_chart(countries, percentages)

#Entry point y que tenga dualidad, con esto se ejecuta desde example y desde main con la terminal
if __name__ == '__main__':
  run()
```
